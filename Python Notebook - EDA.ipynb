```python
from scipy import stats

# Z-score method to handle outliers
threshold = 3
df_zscore = df.copy()

for col in numeric_cols:
    z_scores = stats.zscore(df_zscore[col])
    abs_z_scores = abs(z_scores)
    outliers = (abs_z_scores > threshold)
    df_zscore.loc[outliers, col] = df_zscore[col].mean()

# IQR method to handle outliers
df_iqr = df.copy()

for col in numeric_cols:
    Q1 = df_iqr[col].quantile(0.25)
    Q3 = df_iqr[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = ((df_iqr[col] < lower_bound) | (df_iqr[col] > upper_bound))
    df_iqr.loc[outliers, col] = df_iqr[col].median()

# Checking for remaining outliers
zscore_outliers_count = (abs(stats.zscore(df_zscore[numeric_cols])) > threshold).sum()
iqr_outliers_count = (((df_iqr[numeric_cols] < lower_bound) | (df_iqr[numeric_cols] > upper_bound)).sum())

# Display the counts of outliers detected by each method
outliers_summary = pd.DataFrame({
    'Z-Score Outliers': zscore_outliers_count,
    'IQR Outliers': iqr_outliers_count
})

print("Outliers Summary:")
display(outliers_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame and has been already loaded
# Select categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Determine the number of rows and columns for the grid
n_cols = 3
n_rows = len(categorical_cols) // n_cols + int(len(categorical_cols) % n_cols > 0)

# Create a figure and a grid of subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 5))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# Plot count plots for each categorical column
for i, col in enumerate(categorical_cols):
    ax = axes[i // n_cols, i % n_cols]
    sns.countplot(y=df[col], ax=ax)
    ax.set_title(f'Count Plot of {col}')
    ax.set_ylabel(col)
    ax.set_xlabel('Count')

# Remove any empty subplots
for j in range(i + 1, n_rows * n_cols):
    fig.delaxes(axes[j // n_cols, j % n_cols])

plt.tight_layout()
plt.show()

```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a directory to save the plots if it doesn't exist
plot_dir = "plots"
os.makedirs(plot_dir, exist_ok=True)

# Example plot saving function
def save_plot(plot_func, filename, *args, **kwargs):
    plt.figure(figsize=(10, 5))
    plot_func(*args, **kwargs)
    plt.savefig(os.path.join(plot_dir, filename))
    plt.close()

# Save histograms for numerical columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    save_plot(sns.histplot, f'histogram_{col}.png', df[col].dropna(), kde=True, bins=30)

# Save box plots for numerical columns
for col in numeric_cols:
    save_plot(sns.boxplot, f'boxplot_{col}.png', df[col].dropna())

# Save correlation heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.savefig(os.path.join(plot_dir, 'correlation_heatmap.png'))
plt.close()

# Log transformation for 'Dexa_Freq_During_Rx'
df['Dexa_Freq_During_Rx_log'] = np.log1p(df['Dexa_Freq_During_Rx'])

# Save original and log-transformed histograms
save_plot(plt.hist, 'histogram_Dexa_Freq_During_Rx.png', df['Dexa_Freq_During_Rx'].dropna(), bins=30, edgecolor='black')
save_plot(plt.hist, 'histogram_Dexa_Freq_During_Rx_log.png', df['Dexa_Freq_During_Rx_log'].dropna(), bins=30, edgecolor='black')

# Save original and log-transformed box plots
save_plot(plt.boxplot, 'boxplot_Dexa_Freq_During_Rx.png', df['Dexa_Freq_During_Rx'].dropna(), vert=False)
save_plot(plt.boxplot, 'boxplot_Dexa_Freq_During_Rx_log.png', df['Dexa_Freq_During_Rx_log'].dropna(), vert=False)

# Save pair plot
sns.pairplot(df[numeric_cols])
plt.title('Pair Plot of Numerical Features')
plt.savefig(os.path.join(plot_dir, 'pair_plot.png'))
plt.close()

# Save count plots for categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    save_plot(sns.countplot, f'countplot_{col}.png', y=df[col])

# Save heatmap for categorical data
crosstab = pd.crosstab(df['Gender'], df['Persistency_Flag'])
plt.figure(figsize=(10, 5))
sns.heatmap(crosstab, annot=True, cmap='coolwarm', fmt='d')
plt.title('Heatmap of Gender vs Persistency Flag')
plt.savefig(os.path.join(plot_dir, 'heatmap_Gender_vs_Persistency_Flag.png'))
plt.close()
```
